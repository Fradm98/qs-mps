## Energy level closing for ANNNI Model

```python
from qs_mps.sparse_hamiltonians_and_operators import *

import numpy as np
from matplotlib import pyplot as plt
```

```python
Ls = [8,10,12,14,16]
h = 1e-3
eps = 1e-5
grid = np.linspace(0,1,32)
e_gs = []
e_ex = []
for L in Ls:
    v0 = np.full(shape=(2**L), fill_value=(1/np.sqrt(2))**L)
    e_gs_L = []
    e_ex_L = []
    for k in grid:
        print(f"search for k: {k:.3f}, L:{L}")
        H = ANNNI(L=L, k=k, h=h, eps=eps, deg_method=2)
        e, v = diagonalization(H, sparse=True, v0=v0, k=2)
        v0 = v[:,0]
        e_gs_L.append(e[0])
        e_ex_L.append(e[1])
    e_gs.append(e_gs_L)
    e_ex.append(e_ex_L)
```

```python
for L,gs,ex in zip(Ls, e_gs, e_ex):
    print(np.abs(np.asarray(gs)-np.asarray(ex)))
    plt.plot(grid, np.abs(np.asarray(gs)-np.asarray(ex)), label=f"L={L}")
plt.title("Energy difference between $E_{gs}$ and $E_{ex}$")
plt.yscale('log')
plt.legend()
plt.show()
```
