```python
import os

def rename_files(directory, filename1, filename2):
    for filename in os.listdir(directory):
        # Check if the file is of the type we want to rename and does not already contain 'bc_pbc' or 'bc_obc'
        if filename.startswith(filename1) or filename.startswith(filename2):
            if "_bc_pbc_" in filename or "_bc_obc_" in filename:
                # Skip renaming if 'bc_pbc' or 'bc_obc' is already present
                continue

            # Insert 'bc_obc' if not present
            if "None-None" in filename:
                new_name = filename.replace("None-None", "bc_obc_None-None")
            else:
                new_name = filename.replace("_[", "_bc_obc_[")

            old_file = os.path.join(directory, filename)
            new_file = os.path.join(directory, new_name)
            os.rename(old_file, new_file)
            print(f"Renamed: {filename} to {new_name}")
```
