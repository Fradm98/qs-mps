## Fit of confined and deconfined phases

```python
# interval = np.logspace(-5,1,20)
interval = np.linspace(0,10,40)[:17]
interval = 1/np.sqrt(interval)
delta_energy = np.abs(np.asarray(en_charges)-np.asarray(en_vacuum)).flatten()

def func_conf(x,a,R):
    return a + 2*x*R

def func_deconf(x,a,b,c,R):
    return a*(x**2) + b*np.exp(-R/x) + c

popt_dec, copt_dec = curve_fit(func_deconf, interval[8:16], delta_energy[8:16], [1,1,0,2], maxfev=2000)
print(f"R: {popt_dec[3]}")
popt_conf, copt_conf = curve_fit(func_conf, interval[6:10], delta_energy[6:10], [0,6], maxfev=2000)
print(f"R: {popt_conf[1]}")

plt.plot(interval, delta_energy, 'o', label="energy diff")
plt.plot(interval[8:16], func_deconf(interval[8:16],popt_dec[0],popt_dec[1],popt_dec[2],popt_dec[3]), label="fit deconf")
plt.plot(interval[6:10], func_conf(interval[6:10],popt_conf[0],popt_conf[1]), label="fit conf")
# plt.plot(interval, func(interval,1,1,0,2), label="fit")
#
# plt.yscale('log')
# plt.xscale('log')
# plt.ylim(1e-12,1e-7)
plt.legend()
plt.show()
```
