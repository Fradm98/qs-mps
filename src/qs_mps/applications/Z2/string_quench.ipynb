```python
import matplotlib.pyplot as plt
import numpy as np
from qs_mps.utils import load_list_of_lists, tensor_shapes
from qs_mps.mps_class import MPS
from scipy.optimize import curve_fit
```

```python
en_charges = load_list_of_lists("G:\My Drive\projects\\1_Z2\\results\energy_data\energies_Z2_dual_direct_lattice_4x5_2_particle(s)_sector_[1, 4]-[2, 2]_h_0.001-0.01_delta_10_chi_200")
en_vacuum = load_list_of_lists("G:\My Drive\projects\\1_Z2\\results\energy_data\energies_Z2_dual_direct_lattice_4x5_vacuum_sector_None-None_h_0.001-0.01_delta_10_chi_200")
```

```python
plt.plot(np.abs(en_charges), label="2-particle sector")
plt.plot(np.abs(en_vacuum), label="vacuum sector")
plt.yscale('log')
plt.legend()
plt.show()
```

```python
interval = np.logspace(np.log10(1),np.log10(1e+10),10)
delta_energy = np.abs(np.asarray(en_charges)-np.asarray(en_vacuum)).flatten()

def func(x,a,R):
    return a + 2*x*R

popt, copt = curve_fit(func, interval, delta_energy)
print(f"R: {popt[1]}")

plt.plot(interval, delta_energy, 'o', label="energy diff")
plt.plot(interval, func(interval,popt[0],popt[1]), label="fit")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.show()
```

```python
path_tensor = "/Users/fradm98/Desktop/projects/1_Z2"
charges_x = [1,5]
charges_y = [2,2]
precision = 2
lattice_mps = MPS(L=7,d=16, model="Z2_dual", chi=50, h=10)
lattice_mps.L = lattice_mps.L - 1
lattice_mps.Z2.add_charges(charges_x,charges_y)
lattice_mps.load_sites(path=path_tensor, precision=precision, cx=charges_x, cy=charges_y)
```

```python
E, mag, W, S, err, entr = lattice_mps.TEBD_variational_Z2(trotter_steps=50, delta=0.1, h_ev=1)
```

```python
for t, e_trott in enumerate(E):
    plt.title(f"Trotter step: {t}")
    plt.imshow(e_trott, vmin=-1, vmax=1)
    plt.show()
```
