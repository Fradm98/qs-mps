```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
energies = np.load("G:\My Drive\projects\\3_CLUSTER\\results\energy_data\energies_Cluster_L_12_h_0.0-1.0_delta_3_chi_64.npy")
```

```python
hxs = np.linspace(0,1,3)
hys = np.linspace(0,1,3)
params = [[i,j] for i,hx in enumerate(hxs) for j,hy in enumerate(hys)]
energy_densities = [np.abs(energies[i,j,-1]-energies[i,j])/12 for i,hx in enumerate(hxs) for j,hy in enumerate(hys)]
for i,ed in enumerate(energy_densities):
    plt.plot(ed, label=f"param: {params[i][0]},{params[i][1]}")
plt.legend()
plt.yscale('log')
plt.show()
```
